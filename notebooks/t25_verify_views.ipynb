```python
import os
from pathlib import Path
from dotenv import load_dotenv
from dbrepo.RestClient import RestClient

load_dotenv(dotenv_path=Path('../.env'))
TU_USERNAME = os.getenv('TU_USERNAME')
TU_PASSWORD = os.getenv('TU_PASSWORD')

DATABASE_ID = "ed890fa1-154c-4a66-8529-4088c97f68db"
DBREPO_ENDPOINT = "https://test.dbrepo.tuwien.ac.at"

client = RestClient(
    endpoint=DBREPO_ENDPOINT,
    username=TU_USERNAME,
    password=TU_PASSWORD
)

print("Client ready!")
```

```python
views_to_check = [
    ("view_no2_classification_features", "1467bc6c-b8b0-44fd-8c2a-80209e5ebeb5"),
]

for view_name, view_id in views_to_check:
    try:
        df = client.get_view_data(database_id=DATABASE_ID, view_id=view_id)
        print(f"✓ {view_name}: {len(df)} rows returned")
    except Exception as e:
        print(f"✗ {view_name}: {e}")
```
